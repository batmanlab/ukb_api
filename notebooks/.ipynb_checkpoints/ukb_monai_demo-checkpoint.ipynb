```python
import ukb_api
from ukb_api import module_bulk_data_handler as bdh
import monai
from monai.data import Dataset
from monai.data import NibabelReader
```

```python
print(ukb_api.__version__)
```

```python
ukb_api_object=bdh.bulk_data_handler()
```

```python
ukb_api_object.get_field_ids_for_category("T1 structural brain MRI")
```

```python
#enforce parameter name
t1_subject_list = ukb_api_object.get_subject_list_field_ids(20252)[:10]
```

```python
# we have downloaded t1 data for these subejcts
downloaded_subjects = ['1026859', '1666281','1020970', '2687717','1017426', '1747567', '1226225', '1651490', '1155917', '1925372']
```

```python
t1_file_paths = ukb_api_object.fetch_bulk_data(downloaded_subjects, data_type_string="T1_Image")
```

```python
monai_test_dataset = Dataset(t1_file_paths)
image_reader = NibabelReader()
```

```python
from monai.visualize import blend_images, matshow3d, plot_2d_or_3d_image
```

```python
img_obj = image_reader.read(monai_test_dataset.data[0])
img_data, meta_data = image_reader.get_data(img_obj)
```
