```python
field_ids_df.head(5)
```

```python
# 61 fields marked ‡ are blob/bulk.
# 8 fields marked † are compound.
# 87 fields marked * are view-only.
# 11 fields marked #; are restricted.
field_ids_df['Type'].value_counts()
```

```python
import numpy as np
def assign_type(type):
    
    if type=="‡": return "blob/bulk data"
    
    elif type=="†": return "compound data"
    
    elif type=="*": return "view only data"
    
    elif  type=="#": return "restricted data type"
    
    elif type is np.NaN : return "type not available"
    
       
field_ids_df['field_id_category'] = field_ids_df['Type'].apply(lambda x: assign_type(x))
    
```
