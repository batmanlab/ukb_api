```python
import os
path="/ocean/projects/asc170022p/tighu/ADNI1_freesurfer6_April_2022"
subject_folders = os.listdir(path)
subject_folders.remove("fsaverage")
subject_folders.remove("adni1_freesurfer6_csv_files")
log_files_path=[]
```

```python
for folder in subject_folders:
    log_files_path.append(path+"/"+folder+"/scripts/recon-all.log")
```

```python
log_files_path[:5]
```

```python
len(log_files_path)
```

```python
correct_output_count=0
check_string="finished without error"
currently_at_file=1
images_with_error=[]
for path in log_files_path:

    with open(path, 'rb') as f:
        if currently_at_file % 500 ==0:
            print(currently_at_file)
        f.seek(-2, os.SEEK_END)
        while f.read(1) != b'\n':
            f.seek(-2, os.SEEK_CUR)
        last_line = f.readline().decode()
        if check_string in last_line:
            correct_output_count=correct_output_count+1
        else:
            images_with_error.append(path)
        currently_at_file=currently_at_file+1
print(correct_output_count)
```
